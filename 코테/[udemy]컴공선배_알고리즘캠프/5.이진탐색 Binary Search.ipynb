# 파라메트릭 서치 중요! (코드 문제 다수 출제됨)

- 최적화 문제: 문제 상황을 만족하는 변수의 min, max를 구하는 문제
- 결정 문제: Yes/No

- Ex. 유데미 수강생들의 외모값과 커플/솔로 여부가 주어진다./ 커플은 솔로들보다 외모값이 높다
- Q2. 외모값이 최소 몇 이상일 때부터 커플인가?

- 다른 알고리즘과 융합하는 코드문제들이 많이 나온다. 
- True/ False 를 구성하는 것이 관건인 경우가 많다

### 백준 2512 예산

```python
N = int(input())
req=list(map(int, input().split()))
M=int(input())
ans=0

low=0
hi=max(req)
mid=(lo+hi) // 2

def is_possible(mid):
    return sum(min(r, mid) for r in req) <= M
    

#이분탐색
while lo <= hi:
    if is_possible(mid):
        lo=mid+1 # 탐색범위 줄이기
        ans=mid
    else:
        hi=mid-1
        
    mid=(lo+hi)//2
    


print(ans)
```

### 백준 10815 숫자카드

```python
from bisect import bisect_left, bisect_right

N = int(input())
cards = sorted(map(int, input().split()))
M = int(input())
qry = list(map(int, input().split()))

ans = []

for q in qry:
    l = bisect_left(cards, q)
    r = bisect_right(cards, q)
    ans.append(1 if r - l else 0)
    # if cards[l] == q:

print(*ans)
# print(' '.join(map(str, ans)))
```
