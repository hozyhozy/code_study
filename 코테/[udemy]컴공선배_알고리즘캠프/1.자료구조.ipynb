```python
import math
def solution(progresses, speeds):
    
    DATE=[]
    for i in range(len(progresses)):
        date=math.ceil((100-progresses[i])/speeds[i])
        DATE.append(date)
        
    answer=[]
    num=1
    for d in range(len(DATE)-1):
        if DATE[0] >= DATE[1]:
            DATE.pop(1)
            num+=1
        
        else:
            answer.append(num)
            DATE.pop(0)
            num=0
            num+=1
    answer.append(num) # 마지막 거 append

            
    return answer
```

```python
def solution(progresses, speeds):
    print(progresses)
    print(speeds)
    answer = []
    time = 0
    count = 0
    while len(progresses)> 0: # progess전체 돌리겠다.
        if (progresses[0] + time*speeds[0]) >= 100: # 
            progresses.pop(0)
            speeds.pop(0)
            count += 1
        else:
            if count > 0:
                answer.append(count)
                count = 0
            time += 1
    answer.append(count)
    return answer
```

```python
#boj.kr/9012  ㅡ한줄한줄 print 나오게 하면됨.
# 쌍으로 나오는 문제!! 일반적인 stack에 대한 코딩임!
# pop을 이용해보자
T=int(input())
for _ in range(T):
    stack=[]
    isVPS=True
    for ch in input():
        if ch == '(':
            stack.append(ch)
        else:
            if stack:
                stack.pop()
            else:
                isVPS=False
                break
    if stack:
        isVPS=False
        
    print('YES' if isVPS else 'NO')
```

### QUEUE
- deque, popleft()

```python
# boj.kr/2164
# 배열을 만들어보자 (삽입 삭제 시간복잡도!! 중요함)
# 로직대로 짜면 아마도 시간 복잡도가..높아질 것이다...O(N) N<=500000
#1. 맨 앞의 값을 삭제
#2. 맨 앞의 값을 맨 뒤로 보내기 --> 삭제, 삽입  (3N^2)

# queue O(1) 사용해보자! --> 맨 앞뒤로 빼기 가능
from collections import deque

N=int(input())
dq=deque(range(1, N+1))

while len(dq)>1:
    dq.popleft() #맨 앞에 값을 제거한다
    dq.append(dq.popleft()) # 제거된 후 맨 앞에 값을 제거와 동시에 뒤에 넣는다. (이건 append로)

print(dq.popleft())
```

```python
# 튜플로 묶는 방법
import heapq as hq
import sys

N = int(input())
pq=[]

for _ in range(N):
    x=int(sys.stdin.readline()) #파이썬은 입출력도 시간초과됨.! 
    if x:
        hq.heappush(pq, (abs(x),x))
        
    else: # X==0일떄
        print(hq.heappop(pq)[1] if pq else 0)
    
```

```python
# 다른 방법
min_heap=[] # 양수값은 절대값이 작을수록 작음
max_heap=[] # 음수값은 절댓값이 클수록 작음 #max heap의 경우 -부호를 붙여주기

N = int(input())
for _ in range(N):
    x=int(sys.stdin.readline())
    if x:
        if x>0:
            hq.heappush(min_heap, x)
        else:
            hq.heappush(max_heap, -x)
        
    else:
        if min_heap:
            if max_heap:
                if min_heap[0]<abs(-max_heap[0]):
                    print(hq.heappop(min_heap))

                else:
                    print(-hq.heappop(max_heap)) # maxheap의 제일 큰 값 = 절댓값 제일 작은 값
            else:
                print(hq.heappop(min_heap))
        else:
            if max_heap:
                print(-hq.heappop(max_heap))
            else:
                print(0)
                
```

### 사전

```python
# boj.kr/1302
N=int(input())
dic={}

for _ in range(N):
    book=sys.stdin.readline()
    if book in dic:
        dic[book]+=1
    else:
        dic[book]=1

m=max(dic.values())

candi=[]
for k,v in dic.items():
    if v==m:
        candi.append(k)

print(sorted(candi)[0])
```
