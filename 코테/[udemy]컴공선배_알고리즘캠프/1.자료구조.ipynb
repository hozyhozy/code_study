```python
import math
def solution(progresses, speeds):
    
    DATE=[]
    for i in range(len(progresses)):
        date=math.ceil((100-progresses[i])/speeds[i])
        DATE.append(date)
        
    answer=[]
    num=1
    for d in range(len(DATE)-1):
        if DATE[0] >= DATE[1]:
            DATE.pop(1)
            num+=1
        
        else:
            answer.append(num)
            DATE.pop(0)
            num=0
            num+=1
    answer.append(num) #   append

            
    return answer
```

```python
def solution(progresses, speeds):
    print(progresses)
    print(speeds)
    answer = []
    time = 0
    count = 0
    while len(progresses)> 0: # progess .
        if (progresses[0] + time*speeds[0]) >= 100: # 
            progresses.pop(0)
            speeds.pop(0)
            count += 1
        else:
            if count > 0:
                answer.append(count)
                count = 0
            time += 1
    answer.append(count)
    return answer
```

```python
#boj.kr/9012   print  .
#   !!  stack  !
# pop 
T=int(input())
for _ in range(T):
    stack=[]
    isVPS=True
    for ch in input():
        if ch == '(':
            stack.append(ch)
        else:
            if stack:
                stack.pop()
            else:
                isVPS=False
                break
    if stack:
        isVPS=False
        
    print('YES' if isVPS else 'NO')
```

### QUEUE
- deque, popleft()

```python
# boj.kr/2164
#   (  !! )
#     .. ...O(N) N<=500000
#1.    
#2.       --> ,   (3N^2)

# queue O(1) ! -->    
from collections import deque

N=int(input())
dq=deque(range(1, N+1))

while len(dq)>1:
    dq.popleft() #   
    dq.append(dq.popleft()) #         . ( append)

print(dq.popleft())
```

```python
#   
import heapq as hq
import sys

N = int(input())
pq=[]

for _ in range(N):
    x=int(sys.stdin.readline()) #  .! 
    if x:
        hq.heappush(pq, (abs(x),x))
        
    else: # X==0
        print(hq.heappop(pq)[1] if pq else 0)
    
```

```python
#  
min_heap=[] #    
max_heap=[] #     #max heap  - 

N = int(input())
for _ in range(N):
    x=int(sys.stdin.readline())
    if x:
        if x>0:
            hq.heappush(min_heap, x)
        else:
            hq.heappush(max_heap, -x)
        
    else:
        if min_heap:
            if max_heap:
                if min_heap[0]<abs(-max_heap[0]):
                    print(hq.heappop(min_heap))

                else:
                    print(-hq.heappop(max_heap)) # maxheap    =    
            else:
                print(hq.heappop(min_heap))
        else:
            if max_heap:
                print(-hq.heappop(max_heap))
            else:
                print(0)
                
```

```python
# boj.kr/1302
N=int(input())
dic={}

for _ in range(N):
    book=sys.stdin.readline()
    if book in dic:
        dic[book]+=1
    else:
        dic[book]=1

m=max(dic.values())

candi=[]
for k,v in dic.items():
    if v==m:
        candi.append(k)

print(sorted(candi)[0])
```
