# 동전 거스름돈 문제
- 10,50,100,500원 동전들을 무한하게 갖고 있다.
- 손님에게 800원을 거슬러주려고할 떄 동전을 최고한으로 주는 방법은?
- 규칙> 가장 큰 동전을 먼저 최대한 준다

예외문제!
- 100, 400, 500원 동전들을 무한하게 갖고 있다. 
- 손님에게 800원을 거슬러 주려고 할 떄 동전을 최소한으로 주는 방법은?
- 규칙> 400원 2개 주는게 더 효율적

<span style="color:#2D3748; background-color:#fff5b1;"> **위에 문제들의 차이는?
첫번쨰 문제는 나머지 동전이 500원의 약수임 //
두번째 문제는 나머지 동전이 500원의 약수가 아님**</span>

```python
# boj.kr/11047 동전 0
N, K = map(int, input().split())

coins=[int(input()) for _ in range(N)]
coins.reverse()
ans=0

for coin in coins:
    ans += K//coin
    K %= coin
    
print(ans)
```

```python
# boj.kr/1449 수리공 항승
# 테이프 길이만큼의 구멍은 무시하고 그 이후의 구멍만을 고려한다.

N, L = map(int, input().split())
coord=[False]*1001  # 0부터 1000까지 구멍을 False로 놓기

for h in map(int,input().split()):
    coord[h]=True # 구멍이 난 곳에 Ture 값을 넣는다

ans=0    
x=0
while x< 1001:
    if coord[x]: # 구멍이 있으면
        ans+=1 # 필요한 테이프 count한다
        x+=L #테이프 길이를 넣어줌 --> 테이프 길이 안에 있는 구멍은 count X
    else:
        x+=1
        
print(ans)

```
