### 문제1: N개의 수를 입력 받아서 두 수를 뽑아 합이 가장 클때는?
- 이중 for 문을 보통 쓴다 --> 시간 초과됨
- 정렬을 사용하는 방법이 있을 수 있다. --> 작은 수부터 정렬해서 가장 뒤에 있는 값 가져오기
- 순열을 사용해보자!!

# 순열 (permutation)
- 모든 경우의 수를 순서대로 살펴볼 때 유용하다

```python
from itertools import permutations
for i in range(4):
    a=permutations(["aya", "ye", "woo", "ma"], i)
    print(a)
```

```python
from itertools import permutations
v=[0,1,2,3]

for i in permutations(v,4): # 몇개 뽑을 것인가? 여기서는 4개 뽑기
    print(i)
```

# 조합 combination
- 순서 상관 X

```python
from itertools import combinations
v=[0,1,2,3]

for i in combinations(v,2):
    print(i)
```

```python
# boj.kr/2309 일곱 난쟁이
from itertools import combinations

height=[int(input()) for _ in range(9)]

for combi in combinations(height, 7): #튜플로 나옴
    if sum(combi)==100:
        for height in sorted(combi):
            print(height)
        break #꼭 해줘야함
```
