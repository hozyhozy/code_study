# 깊이 우선 탐색 DFS (Depth Fisrt Search) = 완전탐색
- 스택이나 재귀를 사용해서 구현한다



```python
adj=[[0]*13 for _ in range(13)]
adj[0][1]= adj[0][7]=1
adj[1][2]=adj[1][5] =1

for row in adj:
    print(row)
```

```python
#재귀로 구현을 해보자 

def dfs(now):
    '''
    현재 내가 방문한 노드 번호
    '''
    for nxt in range(13):
        if adj[now][nxt]: # 간선이 존재할때만
            dfs(nxt) # next노드로 탐색을 진행해라
            
dfs(0) # 0부터 넣으면 알아서 간선이 있을 경우 노드를 호출해줌
```

# 너비 우선 탐색(Breadth First Search) = 완전탐색
- DFS 랑 다른 점은 탐색 순서가 다르다는 점!
- 큐를 사용해서 구현한다

ex. 맨 앞에꺼를 꺼내고 지우고 --> 꺼낸 노드와 연결되어있는 아이들을 맨 뒤에 붙인다. --> 이렇게 반복. leaf node까지 도달하게 됨

```python
from collection import deque

def bfs():
    dq=deque() # queue시작하기위해 deque를 만들어야함
    dq.append(0) #root node
    while dq: # 존재하는 동안!
        now=dq.popleft() # 뺀값을 저장 + 뺴버리기
        for nxt in range(13):
            if adj[now][nxt]: #존재한다면!
                dq.append(nxt)

bfs()
```

```python
import sys
sys.setrecursionlimit(10**6) #파이썬은 깊이를 최대 1000밖에 안되기 때문에 recursion Error가 난다면 추가시켜주기

input=sys.stdin.readline # 인풋이 커서 시간 초과 남 --> 코드 추가

N, M = map(int, input().split())
adj = [[0]*N for _ in range(N)]

for _ in range(M):
    a,b = map(lambda x:x-1, map(int, input().split()))
    adj[a][b]= adj[b][a] = 1

chk=[False]*N # check list는 노드 모음
ans=0

def dfs(now):
    for nxt in range(N):
        if adj[now][nxt] and not chk[nxt]: # 값이 있다면/ 0이 아니라면 (조건이 붙지 않으면 크게 존재 유무를 따짐)
            chk[nxt]=True
            dfs(nxt) # 다음 노드와 연결되어있는 노드를 탐색해줘

for i in range(N):
    if not chk[i]: #chek리스트가 비어있다면
        ans +=1
        chk[i]=True
        dfs(i)
        
print(ans)
```
